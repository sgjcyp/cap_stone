```python
gmaps.configure(api_key=gkey.gmapapikey)
```

```python
locations_gb = df_gb_t07[['latitude','longitude']]
locations_rf = df_rf_t07[['latitude','longitude']]
locations_lin = df_lin_t07[['latitude','longitude']]
# new = old[['A', 'C', 'D']].copy()
# locations.shape
```

```python
weights_gb=abs(df_gb_t07.resid)
weights_rf=abs(df_rf_t07.resid)
weights_lin=abs(df_lin_t07.resid)
# weights.head()
```

```python
fig1 = gmaps.figure()
fig1.add_layer(gmaps.heatmap_layer(locations_gb, weights=weights_gb))
fig1
```

```python
fig2 = gmaps.figure()
fig2.add_layer(gmaps.heatmap_layer(locations_rf, weights=weights_rf))
fig2
```

```python
fig3 = gmaps.figure()
fig3.add_layer(gmaps.heatmap_layer(locations_lin, weights=weights_lin))
fig3
```
